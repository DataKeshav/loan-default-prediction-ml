# Column Name Part-1      
1. LoanID     -----        A unique identifier for each loan in the dataset.
2. Age	      -----        The borrower's age at the time the loan was issued.
3. Income	  -----        The borrower's annual income
4. LoanAmoun  -----        The total amount of the loan that the borrower is requesting or has been approved for.
5. CreditScore -----            A numerical representation of the borrower's creditworthiness, typically ranging from 300 to 850.
                           A higher credit score indicates the borrower is more likely to repay the debt.
6. MonthsEmployed  -----	       The number of months the borrower has been employed at their current job or with their current employer.
7. NumCreditLines -----         The total number of active credit lines (e.g., credit cards, loans) the borrower has at the time of applying for the loan.
8. InterestRate	 -----          The annual percentage rate (APR) charged for borrowing the loan amount, usually expressed as a percentage.
9. LoanTerm	   -----            The length of time (in months) over which the loan is to be repaid.
10. DTIRatio	 -----          The Debt-to-Income ratio, which measures the borrower’debt payments relative to their income. 
                           A higher ratio can indicate greater financial stress.



# Column Name Part-2
11. Education     -----          The highest level of education the borrower has completed (e.g., High School, Bachelor’s, Master’s, etc.).
12. EmploymentType  -----	       The type of employment the borrower is engaged in (e.g., Full-Time, Part-Time, Self-Employed, etc.).
13. MaritalStatus   -----	       The marital status of the borrower (e.g., Single, Married, Divorced, etc.).
14. HasMortgage	   -----        An  indicator (e.g., Yes/No) that shows whether the borrower has an existing mortgage on a property.
15. HasDependents  -----	       An  indicator (e.g., Yes/No) that shows whether the borrower has dependents (children, other family members) to support.
16. LoanPurpose	 -----          The primary reason for taking out the loan (e.g., Home Purchase, Debt Consolidation, Education, etc.).
17. HasCoSigner	 -----          An  indicator (e.g., Yes/No) that shows whether the borrower has a co-signer- 
                            -for the loan (someone who agrees to take responsibility if the borrower defaults).
18. Default	   -----            An  indicator (e.g., Yes/No) that shows whether the borrower defaulted on the loan or failed to make timely payments.
19. Loan Date(DD/MM/YYYY)  ----- The date the loan was issued or originated.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check target distribution
print("Target Variable: 'Default'")
print("\nValue counts:")
print(df['Default'].value_counts())

# Visualize
plt.figure(figsize=(10, 5))
df['Default'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Loan Default Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Default Status')
plt.ylabel('Number of Loans')
plt.xticks(rotation=0)
plt.show()

# Calculate default rate
default_rate = (df['Default'] == 1).mean() * 100
    
print(f"\n Default Rate: {default_rate:.2f}%")
```

```python
## Independent and Dependent Feature
X=df.drop('Default',axis=1)
y=df['Default']
```

```python
X.head()
```

```python
y.head()
```

```python
# Encoding categorical values for model.
from sklearn.preprocessing import LabelEncoder

# categorical columns
categorical_cols = ['Education', 'EmploymentType', 'MaritalStatus', 
                   'HasMortgage', 'HasDependents', 'LoanPurpose', 'HasCoSigner']

print("Encoding categorical variables...")
print("="*60)

label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    label_encoders[col] = le
    print(f"Encoded: {col}")

print("\n All categorical variables encoded!")

# Verify everything is numeric now
print(f"\n Data types after encoding:")
print(X.dtypes)
print(f"\n  All {len(X.columns)} features are now numeric!")
```

```python
# Train-test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42,stratify = y)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import time

# Define 3 models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=15, random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42)
}

# Store results
results = {}

print(" TRAINING MACHINE LEARNING MODELS")

for name, model in models.items():
    print(f"\n🔹 {name}...")
    
    start = time.time()
    
    # Train
    model.fit(X_train, y_train)
    
    # Predict
    y_pred = model.predict(X_test)
    
    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    training_time = time.time() - start
    
    # Store
    results[name] = {
        'model': model,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'time': training_time
    }
    
    # Print
    print(f"   Accuracy:  {accuracy*100:.2f}%")
    print(f"   Precision: {precision*100:.2f}%")
    print(f"   Recall:    {recall*100:.2f}%")
    print(f"   F1-Score:  {f1*100:.2f}%")
    print(f"   Time:      {training_time:.2f}s")

print(" ALL MODELS TRAINED!")

```

```python
import pandas as pd

# Comparison table
comparison = pd.DataFrame({
    'Model': list(results.keys()),
    'Accuracy': [results[m]['accuracy'] for m in results],
    'Precision': [results[m]['precision'] for m in results],
    'Recall': [results[m]['recall'] for m in results],
    'F1-Score': [results[m]['f1_score'] for m in results]
}).round(4)

print("\n MODEL PERFORMANCE COMPARISON")
print(comparison.to_string(index=False))

# Best model
best = comparison.loc[comparison['F1-Score'].idxmax(), 'Model']
best_f1 = comparison['F1-Score'].max()

print(f"\n  BEST MODEL: {best}")
print(f"   F1-Score: {best_f1*100:.2f}%")

```

```python
gb_model = results['Gradient Boosting']['model']
y_pred_gb = results['Gradient Boosting']['model'].predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred_gb)
print("Confusion matrix:\n", cm)

print("\nClassification report:")
print(classification_report(y_test, y_pred_gb, target_names=['No Default', 'Default']))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print(" Logistic Regression with class_weight='balanced'")

# 1. Create model with balanced class weights
bal_lr = LogisticRegression(
    max_iter=1000,
    random_state=42,
    class_weight='balanced'   
)

# 2. Train
bal_lr.fit(X_train, y_train)

# 3. Predict
y_pred_bal_lr = bal_lr.predict(X_test)

# 4. Evaluate
acc = accuracy_score(y_test, y_pred_bal_lr)
cm = confusion_matrix(y_test, y_pred_bal_lr)

print(f"\nAccuracy: {acc*100:.2f}%")
print("\nConfusion matrix:")
print(cm)

print("\nClassification report:")
print(classification_report(y_test, y_pred_bal_lr,target_names=['No Default', 'Default']))
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Get predicted probabilities for class 1 (Default)
y_proba = bal_lr.predict_proba(X_test)[:, 1]

for thresh in [0.5, 0.4, 0.3]:
    print(f"\n Threshold: {thresh} ")
    y_pred_thresh = (y_proba >= thresh).astype(int)

    acc = accuracy_score(y_test, y_pred_thresh)
    cm = confusion_matrix(y_test, y_pred_thresh)

    print(f"Accuracy: {acc*100:.2f}%")
    print("Confusion matrix:")
    print(cm)

    print("\nClassification report:")
    print(classification_report(y_test, y_pred_thresh, target_names=['No Default', 'Default']))
```
